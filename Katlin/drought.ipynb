```python
#Import dependencies
import pandas as pd
from census import Census
import matplotlib.pyplot as plt
import scipy.stats as st

# Census & gmaps API Keys
from config import (api_key)
```

```python
c = Census(api_key, year=2013)

# Run Census Search to retrieve data on all states
census_data = c.acs5.get(("NAME", "B19013_001E", "B01003_001E", "B01002_001E",
                          "B19301_001E",
                          "B17001_002E",
                          "B23025_005E"), {'for':'county:*','in':'state:06'})

# Convert to DataFrame
census_pd = pd.DataFrame(census_data)

# Column Reordering
census_pd = census_pd.rename(columns={"B01003_001E": "Population",
                                      "B01002_001E": "Median Age",
                                      "B19013_001E": "Household Income",
                                      "B19301_001E": "Per Capita Income",
                                      "B17001_002E": "Poverty Count",
                                      "B23025_005E": "Unemployment Count",
                                      "NAME": "Name", "state": "State"})
                                           
#Split NAME column to separate county and state
census = census_pd["Name"].astype(str)
census = census.str.split(pat=",")
data = census.tolist()
county_state = ["County name", "State",]
census = pd.DataFrame(data, columns = county_state)

census_pd['County'] = census['County name']

# Add in Poverty Rate (Poverty Count / Population)
census_pd["Poverty Rate"] = 100 * \
    census_pd["Poverty Count"].astype(
        int) / census_pd["Population"].astype(int)

#Add year column

census_pd['Year'] = '2013'

census_2013 = census_pd
```

```python
#Clean drought data
#Import CSV file 
csv_path = "dm_export_2009_2019.csv"
drought_data = pd.read_csv(csv_path)
drought = pd.DataFrame(drought_data)
drought.head()
```

```python
#Split ValidEnd column to separate year
drought_date = drought["ValidEnd"].astype(str)
drought_date_split = drought_date.str.split(pat="/")
data = drought_date_split.tolist()
names = ["Month", "Day", "Year"]
year_df = pd.DataFrame(data, columns = names)
year_df

#Add year column to original dataframe
drought["Year"] = year_df["Year"]
drought.head()
```

```python
#Group df by year and county, and take the average of each drought index column
drought_filtered = drought.groupby(['Year', 'County']).mean()
drought_filtered.head()
drought_ave = pd.DataFrame(drought_filtered)
drought_ave.head()
```

```python
#Reset index, match Year and County datatypes to census dataframe
drought_df = drought_ave.reset_index()
drought_df = drought_df.astype({'Year':'int64'})
drought_df.dtypes
```

```python
#Concatenate census data from 2013-2019
census_data = pd.concat([census_2013,census_2014, census_2015,census_2016,census_2017,census_2018,census_2019], axis=0)
census_data.dtypes
```

```python
#Merge drought data with Kevin's census data

#Import ten years of census data
csv_path = "Kevin/Census_10_19.csv"
census_10 = pd.read_csv(csv_path)
census_10 = pd.DataFrame(census_10)
census_10.head()
census_10.dtypes
```

```python
#Merge census and drought data, drop NA values 
merged_df = pd.merge(drought_df, census_10, how="outer", on = ['Year', 'County'])
merged_df = merged_df.drop(columns=['MapDate', 'FIPS','State', 'StatisticFormatID'])
merged_df = merged_df.loc[merged_df['Year']>2009]
merged_df = merged_df.loc[merged_df['Year']<2020]
merged_df.dropna()
merged_df.head()
```

```python
#Save merged dataframe as csv
merged_df.to_csv('drought_census_2010_2019.csv')
merged_df.head()
```

```python
# Create a scatterplot for poverty rate v. D4 

#Define variables
x = merged_df['Poverty Rate']
y = merged_df['D4']

#Plot data
plt.figure(figsize=(8,8))
plt.title("Poverty Rate v. Exceptional Drought (2010-2019)", fontsize = '18')
plt.xlabel("Poverty Rate", fontsize = '18')
plt.ylabel("Population in Exceptional Drought (%)", fontsize = '18')
plt.xticks(fontsize=15)
plt.yticks(fontsize=15)
cap_plot = plt.scatter(x,y, marker="o", color="red",edgecolors="black",s=60)
plt.show()
```

```python

# Create a scatterplot for poverty rate v. D4 >75% 
D4 = merged_df.loc[merged_df['D4']>=75]

#Define variables
x = D4['Poverty Rate']
y = D4['D4']

#Plot data
plt.figure(figsize=(8,8))
plt.title("Poverty Rate v. Exceptional Drought (2010-2019)", fontsize = '18')
plt.xlabel("Poverty Rate", fontsize = '18')
plt.ylabel("Population in Exceptional Drought (%)", fontsize = '18')
plt.xticks(fontsize=15)
plt.yticks(fontsize=15)
cap_plot = plt.scatter(x,y, marker="o", color="red",edgecolors="black", s=60)


(slope, intercept, rvalue, pvalue, stderr) = st.linregress(x, y)
regress_values = x * slope + intercept
line_eq = "y = " + str(round(slope,2)) + "x + " + str(round(intercept,2))

print(f'Correlation coefficient = {round(rvalue,2)}')
print(f'Linear regression model: {line_eq}')
print(f'P-Value = {pvalue}')

plt.plot(x,regress_values,"black")
plt.annotate(line_eq,(6,10),fontsize=15,color="black")


plt.show()
```
