```python
#Import dependencies
import pandas as pd
from census import Census

# Census & gmaps API Keys
from config import (api_key)
```

```python
c = Census(api_key, year=2013)

# Run Census Search to retrieve data on all states
census_data = c.acs5.get(("NAME", "B19013_001E", "B01003_001E", "B01002_001E",
                          "B19301_001E",
                          "B17001_002E",
                          "B23025_005E"), {'for':'county:*','in':'state:06'})

# Convert to DataFrame
census_pd = pd.DataFrame(census_data)

# Column Reordering
census_pd = census_pd.rename(columns={"B01003_001E": "Population",
                                      "B01002_001E": "Median Age",
                                      "B19013_001E": "Household Income",
                                      "B19301_001E": "Per Capita Income",
                                      "B17001_002E": "Poverty Count",
                                      "B23025_005E": "Unemployment Count",
                                      "NAME": "Name", "state": "State"})
                                           
#Split NAME column to separate county and state
census = census_pd["Name"].astype(str)
census = census.str.split(pat=",")
data = census.tolist()
county_state = ["County name", "State",]
census = pd.DataFrame(data, columns = county_state)

census_pd['County'] = census['County name']

# Add in Poverty Rate (Poverty Count / Population)
census_pd["Poverty Rate"] = 100 * \
    census_pd["Poverty Count"].astype(
        int) / census_pd["Population"].astype(int)

#Add year column

census_pd['Year'] = '2013'

census_2013 = census_pd
```

```python
#Clean drought data
#Import CSV file 
csv_path = "drought_export"
drought_data = pd.read_csv('drought_export.csv')
drought = pd.DataFrame(drought_data)
drought.head()
```

```python
#Split ValidEnd column to separate year
drought_date = drought["ValidEnd"].astype(str)
drought_date_split = drought_date.str.split(pat="/")
data = drought_date_split.tolist()
names = ["Month", "Day", "Year"]
year_df = pd.DataFrame(data, columns = names)
year_df

#Add year column to original dataframe
drought["Year"] = year_df["Year"]
drought.head()
```

```python
#Group df by year and county, and take the average of each drought index column
drought_filtered = drought.groupby(['Year', 'County']).mean()
drought_filtered.head()
drought_ave = pd.DataFrame(drought_filtered)
drought_ave.head()
```

```python
#Reset index, match Year and County datatypes to census dataframe
drought_df = drought_ave.reset_index()
drought_df.astype({'Year':'int64'}).dtypes
drought_df
```

```python
#Concatenate census data from 2013-2019
census_data = pd.concat([census_2013,census_2014, census_2015,census_2016,census_2017,census_2018,census_2019], axis=0)
census_data.head()
```

```python
#Merge census and drought data
merged_df = pd.merge(drought_df, census_data, how="outer", on = ['Year', 'County'])
merged_df.head()
merged_df.drop(columns=['MapDate', 'FIPS', 'Name','State','county','StatisticFormatID'])
```

```python
#Save merged dataframe as csv
merged_df.to_csv('drought_census.csv')
```
