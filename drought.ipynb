```python
#Import dependencies
import pandas as pd
```

```python
#Import CSV file 
csv_path = "drought_export"
drought_data = pd.read_csv('drought_export.csv')
drought = pd.DataFrame(drought_data)
drought.head()
```

```python
#Split ValidEnd column to separat year
drought_date = drought["ValidEnd"].astype(str)
drought_date_split = drought_date.str.split(pat="/")
data = drought_date_split.tolist()
names = ["Month", "Day", "Year"]
year_df = pd.DataFrame(data, columns = names)
year_df

#Add year column to original dataframe
drought["Year"] = year_df["Year"]
drought.head()
```

```python
#Group df by year and county, and take the average of each drought index column
drought_filtered = drought.groupby(['Year', 'County'])
drought_filtered.head()
```

```python
#Calculate the average for each drought index
none_ave = drought_filtered["None"].mean()
D0_ave = drought_filtered["D0"].mean()
D1_ave = drought_filtered["D1"].mean()
D2_ave = drought_filtered["D2"].mean()
D3_ave = drought_filtered["D3"].mean()
D4_ave = drought_filtered["D4"].mean()

#Create new dataframe with average values
D4_ave
```
