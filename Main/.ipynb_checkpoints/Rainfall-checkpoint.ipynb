```python
import pandas as pd
```

```python
path = ('Resources/Rain/pcp_all.csv')
data = pd.read_csv(path)
data
```

```python
data['Year'] = data['Date'].astype(str).str.extract('(\d{,4})', expand=False)
data['Month'] = data['Date'].astype(str).str.strip().str[-2:]
data.head()
```

```python
data = data[['Location ID', 'Location', 'Year', 'Month', 'Value']]
data = data.rename(columns = {'Location ID' : 'State-FIP', 'Location' : 'County'})
```

```python
data = data.loc[(data['Year'] >= '2010' & (data['Year'] <= '2019')]
```

```python
data_year = data.groupby('Year')['Value'].mean()
```
