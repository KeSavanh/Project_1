# Census Data cleaning

```python
# Retrive census data for 2010
c = Census(api_key, year=2010)
# Run Census Search to retrieve data on all states
census_data = c.acs5.get(("NAME", "B19013_001E", "B01003_001E", "B01002_001E",
                          "B19301_001E",
                          "B17001_002E"), {'for':'county:*','in':'state:06'})
# Convert to DataFrame
census_pd = pd.DataFrame(census_data)

# Column Reordering
census_pd = census_pd.rename(columns={"B01003_001E": "Population",
                                      "B01002_001E": "Median Age",
                                      "B19013_001E": "Household Income",
                                      "B19301_001E": "Per Capita Income",
                                      "B17001_002E": "Poverty Count",
                                      "NAME": "Name", "state": "State"})
                                           
#Split NAME column to separate county and state
census = census_pd["Name"].astype(str)
census = census.str.split(pat=",")
data = census.tolist()
county_state = ["County name", "State",]
census = pd.DataFrame(data, columns = county_state)

census_pd['County'] = census['County name']

# Add in Poverty Rate (Poverty Count / Population)
census_pd["Poverty Rate"] = 100 * \
    census_pd["Poverty Count"].astype(
        int) / census_pd["Population"].astype(int)

#Add year column
census_pd['Year'] = '2010'
census_2010 = census_pd
```

# Drought data cleaning

```python
#Import CSV file 
csv_path = "Resources/dm_export_2009_2019.csv"
drought_data = pd.read_csv(csv_path)
drought = pd.DataFrame(drought_data)
drought.head()
```

```python
#Split ValidEnd column to separate year
drought_date = drought["ValidEnd"].astype(str)
drought_date_split = drought_date.str.split(pat="/")
data = drought_date_split.tolist()
names = ["Month", "Day", "Year"]
year_df = pd.DataFrame(data, columns = names)
year_df

#Add year column to original dataframe
drought["Year"] = year_df["Year"]
drought.head()
```

```python
#Group df by year and county, and take the average of each drought index column
drought_filtered = drought.groupby(['Year', 'County']).mean()
drought_filtered.head()
drought_ave = pd.DataFrame(drought_filtered)
drought_ave.head()
```

```python
#Reset index, match Year and County datatypes to census dataframe
drought_df = drought_ave.reset_index()
drought_df = drought_df.astype({'Year':'int64'})
drought_df.dtypes
```

```python
#Merge drought data with Kevin's census data

#Import ten years of census data
csv_path = "data_output/Census_10_19.csv"
census_10 = pd.read_csv(csv_path)
census_10 = pd.DataFrame(census_10)
census_10.head()
census_10.dtypes
```

```python
#Merge census and drought data, drop NA values 
merged_df = pd.merge(drought_df, census_10, how="outer", on = ['Year', 'County'])
merged_df = merged_df.drop(columns=['MapDate', 'FIPS','State', 'StatisticFormatID'])
merged_df = merged_df.loc[merged_df['Year']>2009]
merged_df = merged_df.loc[merged_df['Year']<2020]
merged_df.dropna()
merged_df.head()
```

```python
#Save merged dataframe as csv
merged_df.to_csv('data_output/drought_census_2010_2019.csv')
merged_df.head()
```

# Temperature data cleaning

```python
#Import merged dataset 
csv_path = "Resources/max_temp_ca.csv"
df = pd.read_csv(csv_path)
df = pd.DataFrame(df)
df=df.rename(columns={"Location": "County"})
df.head()
```

```python
#sort dataset for 2010-2019
df = df.loc[df["Year"] > 2009]
df = df.loc[df["Year"] <2020]
```

```python
#group by year and location
df.head()
df=df.groupby(['Year', 'County']).max()
```

```python
#Import merged dataset 
csv_path_drought = "data_output/drought_census_2010_2019.csv"
drought_df = pd.read_csv(csv_path_drought)
drought_df = pd.DataFrame(drought_df)
drought_df.head()
```

```python
#Merge census and temp data
merged_df = pd.merge(drought_df, df, how="outer", on = ['Year', 'County'])
merged_df.reset_index()
```

```python
county = merged_df['County'].tolist()
county = [x.rstrip(' County') for x in county]
county
merged_df["County"] = county
merged_df.head()
#Save merged dataframe as csv
merged_df.to_csv('data_output/drought_temp_census_2010_2019.csv')
```

# Wildfire data cleaning

```python
# Load in the csv
wildfire_data = "Resources/California_Fire_Incidents.csv"
```

```python
# Read the wildfire csv and display
fire_data = pd.read_csv(wildfire_data)
fire_data.head()
```

```python
wildfire_df = fire_data[['ArchiveYear', 'Counties', 'AcresBurned']]
wildfire_df.head()
```

```python
# Create data frame from csv data

fireyear = wildfire_df['ArchiveYear']
acres = wildfire_df['AcresBurned']
county = wildfire_df['Counties']

wildfire_df = pd.DataFrame({
    'Year' : fireyear,
    'County' : county,
    'Acres Burned' : acres
})

wildfire_df
```

```python
grouped_wildfire = wildfire_df.groupby(['Year', "County"]).mean()
grouped_wildfire.reset_index()
```

```python
sum_wildfire = wildfire_df.groupby(['Year', "County"]).sum()
sum_wildfire.reset_index()
```

```python
census_csv = 'data_output/drought_census_2010_2019.csv'

census_data = pd.read_csv(census_csv)

census_data.head()
```

```python
# Create data frame from csv census

census_year = census_data['Year']
census_county = census_data['County']
census_poverty = census_data['Poverty Rate']
census_household = census_data['Household Income']

census_df = pd.DataFrame({
    'Year' : census_year,
    'County' : census_county,
    'Poverty Rate' : census_poverty,
    'Household Income' : census_household
})

census_df.head()
```

```python
# Remove the word 'County' from the County column

county_list = census_df.County.tolist()
county_list = [x.rstrip(' County') for x in county_list]
census_df.County = county_list
census_df.head()
```

```python
census_df = census_df.loc[census_df['Year'] > 2012]
census_df = census_df.loc[census_df['Year'] < 2020]
census_df
```

```python
merged_sum_df = pd.merge(census_df, sum_wildfire, how = 'outer', on = ['Year', 'County'], sort = True)
cleaned_merged_sum_df = merged_sum_df.dropna()
cleaned_merged_sum_df
```

```python
cleaned_merged_sum_df.to_csv('data_output/wildfire_sum_census.csv', index = False)
```

```python
# Merge two data frames

merged_avg_df = pd.merge(census_df, grouped_wildfire, how = 'outer', on = ['Year', 'County'], sort = True)
cleaned_merged_avg_df = merged_avg_df.dropna()
cleaned_merged_avg_df
```
